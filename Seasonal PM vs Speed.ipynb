```python
import pandas as pd
traf1 = pd.read_csv('Traffic21.csv')
```

```python
import datetime as dt
traf1['Date'] = pd.to_datetime(traf1['Date'])
```

```python
traf1.dtypes
```

```python
traf = traf1.loc[traf1.CONC != 0.0 ,:]
```

```python
import datetime
t1= pd.to_datetime('1-2-2017')
t2= pd.to_datetime('5-15-2017')
t3= pd.to_datetime('8-1-2017')
t4= pd.to_datetime('11-20-2017')
t5 = pd.to_datetime('11-27-2017')
t6 = pd.to_datetime('12-20-2017')
t7 = pd.to_datetime('1-2-2018')
t8 = pd.to_datetime('5-15-2018')
t9 = pd.to_datetime('8-1-2018')
t10 = pd.to_datetime('11-20-2018')
t11 = pd.to_datetime('11-27-2018')
t12 = pd.to_datetime('12-20-2018')
t13 = pd.to_datetime('1-2-2019')
t14 = pd.to_datetime('5-15-2019')
```

```python
spring1 = traf[traf["Date"].between(t1,t2)]
summer1 = traf[traf["Date"].between(t2,t3)]
fall1 = traf[traf["Date"].between(t3,t4)]
fall2 = traf[traf["Date"].between(t5,t6)]
winter1 = traf[traf["Date"].between(t6,t7)]
spring2 = traf[traf["Date"].between(t7,t8)]
summer2 = traf[traf["Date"].between(t8,t9)]
fall3 = traf[traf["Date"].between(t9,t10)]
fall4 = traf[traf["Date"].between(t11,t12)]
winter2 = traf[traf["Date"].between(t12,t13)]
spring3 = traf[traf["Date"].between(t13,t14)]
```

```python
spring2 = spring2.append(spring3)
spring = spring1.append(spring2)
fall3 = fall3.append(fall4)
fall2 = fall2.append(fall3)
fall = fall1.append(fall2)
winter = winter1.append(winter2)
summer = summer1.append(summer2)
```
